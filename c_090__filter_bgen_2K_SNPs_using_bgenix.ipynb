```bash
# list variants in the chr 22 bgen file
bgenix -g ukb24309_c22_b0_v1.bgen -list > chr22_variants.txt
```

```bash
# check on the file 
# see if it has a header?
head chr22_variants.txt
```

```bash
# randomly select 2K rows of the output file, selecting 2K variants
shuf -n 2000 chr22_variants.txt > chr22_variants_2k.txt
```

```bash
head chr22_variants_2k.txt
```

```bash
# filter bgen file for this 2K variants
bgenix -g ukb24309_c22_b0_v1.bgen -incl-rsids chr22_variants_2k.txt > wgs_chr22_filtered_2k.bgen
```
